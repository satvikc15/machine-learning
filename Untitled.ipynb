```python
import pandas as pd

# Load the dataset
data = pd.read_csv(r'C:\Users\satvi\Downloads\ev_charging_patterns.csv')


# Display basic information about the dataset
data_info = {
    "head": data.head(),
    "description": data.describe(),
    "missing_values": data.isnull().sum(),
    "shape": data.shape
}

data_info
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Selecting numerical features for clustering
numerical_features = [
    "Battery Capacity (kWh)",
    "Energy Consumed (kWh)",
    "Charging Duration (hours)",
    "Charging Rate (kW)",
    "State of Charge (Start %)",
    "State of Charge (End %)",
    "Distance Driven (since last charge) (km)",
    "Temperature (째C)"
]

# Handling missing values with mean imputation
imputer = SimpleImputer(strategy="mean")
data_imputed = imputer.fit_transform(data[numerical_features])

# Standardizing the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_imputed)

# Checking the shape of the processed data
data_scaled.shape
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Applying DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=10)
dbscan_labels = dbscan.fit_predict(data_scaled)

# Checking the number of clusters (excluding noise labeled as -1)
n_clusters_dbscan = len(set(dbscan_labels)) - (1 if -1 in dbscan_labels else 0)
n_noise_dbscan = np.sum(dbscan_labels == -1)

n_clusters_dbscan, n_noise_dbscan
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of the clusters
plt.figure(figsize=(10, 6))

# Assign colors to clusters and noise
unique_labels = set(dbscan_labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for label, color in zip(unique_labels, colors):
    if label == -1:  # Noise
        color = [0, 0, 0, 1]  # Black
    mask = (dbscan_labels == label)
    plt.scatter(data_scaled[mask, 0], data_scaled[mask, 1], c=[color], label=f"Cluster {label}" if label != -1 else "Noise", s=10)

# Adding labels and legend
plt.title(f"DBSCAN Clustering (Clusters: {n_clusters_dbscan}, Noise Points: {n_noise_dbscan})")
plt.xlabel("Feature 1 (Standardized)")
plt.ylabel("Feature 2 (Standardized)")
plt.legend(markerscale=2, bbox_to_anchor=(1.05, 1), loc='upper left', title="Clusters")
plt.grid(True)
plt.tight_layout()
plt.show()
```
